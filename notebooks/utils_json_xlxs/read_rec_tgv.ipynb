```python
import openpyxl

def parse_excel(filename):
    """
    Parse Excel file and return a list of strings.
    """
    workbook = openpyxl.load_workbook(filename)
    
    sheetnames = ["Config-Q", "Config-T", "Config-C"]
    
    str_list = []
    for sheetname in sheetnames:
        current_sheet = workbook[sheetname]

        for row in current_sheet.iter_rows():
            for cell in row:
                if cell.value is not None:
                    str_list.append(str(cell.value))

    return str_list
```

```python
#xlsx_filename = "REC_data_analyst.xlsx"
#xlsx_filename = "REC-charge_etudes.xlsx"
xlsx_filename = "REC-Naomi.xlsx"

list_of_strings = parse_excel(xlsx_filename)

for x in list_of_strings:
    print(x)
```
