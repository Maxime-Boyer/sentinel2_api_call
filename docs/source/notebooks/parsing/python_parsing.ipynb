```python
def parse_pdf(file_path):
    document = fitz.open(file_path)
    text = ""
    page_cmpt = 1

    for page in document:
        text += f"NUMERO DE LA PAGE : {page_cmpt}"
        text += page.get_text()
        page_cmpt += 1

    document.close()
    return text
```

```python
import docx

def read_text_docx(file_path):
    """Read all text from a word (docs) file.

    Args:
    ----
    file_path (str): The path of the word file.

    Returns:
    -------
    str: The extracted text from the word file.

    Dependencies:
    import docx
    """
    # load document
    doc = docx.Document(file_path)
    
    # instantiate return
    full_text = ""
    # parse through paragraphs
    for paragraph in doc.paragraphs:
        full_text += (paragraph.text.strip() + '\n')

    # parse through tables
    for table in doc.tables:
        for row in table.rows:
            for cell in row.cells:
                for paragraph in cell.paragraphs:
                    full_text += (paragraph.text.strip() + '\n')

    # add linebreaks to each elements of the line
    #full_text = [(line + '\n') for line in full_text]

    # join each line
    #string_text = "".join(full_text)

    # remove non ascci character
    string_text = full_text.encode("latin-1", errors="ignore").decode("latin-1")

    # remove trailing spaces and linebreaks
    # string_text = string_text.rstrip("\n")
    # string_text = string_text.strip()

    # return only if not empty
    if string_text:
        return string_text
    else:
        return None
```

```python
def get_all_texts(doc):
    all_texts = []

    # Retrieve text from paragraphs
    for para in doc.paragraphs:
        all_texts.append(para.text)

    # Retrieve text from tables
    for table in doc.tables:
        all_texts.append(get_table_text(table))

    # Retrieve text from shapes in the document XML
    for shape in doc.element.xpath('.//w:drawing//w:t'):
        all_texts.append(shape.text)

    return all_texts

# Get all texts from the document
texts = get_all_texts(doc)

# Print the collected texts
for text in texts:
    print(text)
```

```python
from docx import Document

def get_all_text_in_order(doc):
    all_texts = []
    # This XPath will select paragraph text, table cell content, and shape text
    # as they appear in the XML structure
    xpath_query = './/*[self::w:p or self::w:tc or self::w:drawing]'

    for element in doc.element.xpath(xpath_query):
        if element.tag.endswith('p'):  # Paragraph
            all_texts.append(''.join(node.text for node in element.iterdescendants() if node.tag.endswith('t')))
        elif element.tag.endswith('tc'):  # Table Cell
            texts = [node.text for node in element.iterdescendants() if node.tag.endswith('t')]
            all_texts.append(' '.join(texts).strip())
        elif element.tag.endswith('drawing'):  # Shape
            texts = [node.text for node in element.iterdescendants() if node.tag.endswith('t')]
            all_texts.append(' '.join(texts).strip())

    return all_texts

# Get all texts in order
texts = get_all_text_in_order(doc)

# Print the collected texts
for text in texts:
    print(text)
```

```python
from docx import Document
from docx.text.paragraph import Paragraph
from docx.table import _Cell
from lxml.etree import _Element

class DocElement:
    def __init__(self, element):
        self.element = element
        if isinstance(element, Paragraph):
            self.element_type = 'paragraph'
        elif isinstance(element, _Cell):
            self.element_type = 'table_cell'
        elif isinstance(element, _Element):  # Simplified assumption for drawing
            self.element_type = 'shape'
        else:
            self.element_type = 'unknown'

    def get_text(self):
        if self.element_type == 'paragraph':
            return self.element.text
        elif self.element_type == 'table_cell':
            return ' '.join(p.text for p in self.element.paragraphs if p.text)
        elif self.element_type == 'shape':
            return ""
            #return ''.join(node.text for node in self.element.iterdescendants() if node.tag.endswith('t'))
        return ""

    def __str__(self):
        return f"{self.element_type}: {self.get_text()}"

# Load your document
doc = Document(cv)

# XPath to fetch paragraphs, table cells, and shapes in order
elements = doc.element.xpath('.//*[self::w:p or self::w:tc or self::w:drawing]')

doc_elements = [DocElement(elem) for elem in elements]
# Example usage: print all elements in the document in the order they appear
for elem in doc_elements:
    print(elem.__str__())
```

```python
def search_string(sheet: Worksheet, target: str):
    if not isinstance(target, type(str)): 
        return []
    # kill cases to facilitate quick search via string equality
    target = target.lower()
    # future stores
    locations_and_values = []
    # iterating through rows and columns
    for row in sheet.iter_rows():
        for cell in row:
            # if the current observation is a string, we investigate it
            if isinstance(cell.value, type(str)):
                # kill cases to facilitate quick search via string equality
                current_observation = cell.value.lower()
                # investigate and store if success
                if current_observation == target:
                    locations_and_values.append({"row": cell.row,
                                                 "column": cell.column,
                                                 "value": cell.value})
    return locations_and_values
```

```python
def search_string(sheet: Worksheet, targets):
    if not isinstance(targets, type(list(str))): 
            return []
    
    locations_and_values = []

    for target in targets:
        # kill cases to facilitate quick search via string equality
        target = target.lower()
        # iterating through rows and columns
        for row in sheet.iter_rows():
            for cell in row:
                # if the current observation is a string, we investigate it
                if isinstance(cell.value, type(str)):
                    # kill cases to facilitate quick search via string equality
                    current_observation = cell.value.lower()
                    # investigate and store if success
                    if target in current_observation:
                        locations_and_values.append({
                            "value": cell.value,
                            "row": cell.row,
                            "column": cell.column,
                        })
    
    return locations_and_values
```

```python
strings_1 = [
    "taux journaliers",
    "tranche ferme",
    "abracadabra",
    "tranche optionnelle"
]

locations_values_1 = search_string(sheet_1, strings_1)
locations_values_1
```

```python
infos1 = {}
for lc in locations_values_1:
    if 'taux' in lc["value"].lower():
        row = lc["row"] + 1
        # get profile category value
        profil_column = lc["column"] - 1
        profil_value = sheet_1.cell(row, profil_column).value
        infos1["Catégorie de profil"] = profil_value
        # get daily cost value
        tjm_column = lc["column"]
        tjm_value = sheet_1.cell(row, tjm_column).value
        infos1[lc["value"]] = tjm_value

    elif 'tranche' in lc["value"].lower():
        row = lc["row"] + 2
        column = lc["column"] + 1
        v = sheet_1.cell(row, column).value
        infos1["Nb de jours" + lc["value"]] = v

infos1
```
