```python
import csv

def flatten_value(value, page=""):
    print("Value:", value)
    if isinstance(value, list):
        for item in value:
            print("Item in list:", item)
            yield from flatten_value(item, page)
    elif isinstance(value, dict):
        sub_page = value.get("page", page)
        print("Sub page:", sub_page)
        for key, sub_value in value.items():
            if key != "page":
                print("Key:", key)
                print("Sub value:", sub_value)
                yield from flatten_value(sub_value, sub_page)
    else:
        print("Leaf value:", value)
        yield value, page

create_csv(response_exp, "CV_Exp_json_data.csv")
create_csv(response_tasks, "CV_Tache_json_data.csv")
```
