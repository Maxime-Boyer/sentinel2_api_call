```python
import json
import pandas as pd
import toml
```

```python
config = toml.load("ao_config.toml")

xlsx_outname = f"{config['tender']['id']}.xlsx"
txt_outname = f"{config['tender']['id']}.txt"

# AO SNCF
rec_json_name = config["tender"]["sncf"]["rec"]
cdc_json_name = config["tender"]["sncf"]["cdc"]

# Soumissionnaires
cv_exp_json_name = config["tender"]["responses"]["cv_experiences"]
cv_taches_json_name = config["tender"]["responses"]["cv_taches"]
cv_diplomes_json_name = config["tender"]["responses"]["cv_diplomes"]
```

```python
def _load_json(filename):

    # Opening JSON file
    f = open(filename)
 
    # returns JSON object as a dictionary
    data = json.load(f)

    # Closing file
    f.close()
    
    return data
```

# REC

```python
rec_json_string = _load_json(rec_json_name)

print(json.dumps(rec_json_string, indent=4))
```

```python
def rec_read(filename):
    json = _load_json(filename)

    data = []
    k = "numero_dossier_consultation"
    data.append(f"{k}: {json[k]}")

    k = "nombre_de_profils"
    data.append(f"{k}: {json[k]}")

    k = "types_de_profils"
    data.append(f"{k}: {json[k]}")

    k = "types_de_profils_variante"
    data.append(f"{k}: {json[k]}")
    
    k = "offre_technique"
    tmp = json[k]
    data.append(f"{k}")
    for v in tmp:
        for k1, v1 in v.items():
            data.append(f"\t{k1} : {v1}")

    k = "sous_criteres_techniques_et_ponderation"
    tmp = json[k]
    data.append(f"{k}")
    for v in tmp:
        for k1, v1 in v.items():
            data.append(f"\t{k1} : {v1}")

    k = "note_eliminatoire"
    data.append(f"{k}: {json[k]}")
    
    return data
```

```python
rec_data = rec_read(rec_json_name)

for s in rec_data:
    print(s)
```

# CDC

```python
cdc_json_string = _load_json(cdc_json_name)

print(json.dumps(cdc_json_string, indent=4))
```

```python
def cdc_read(filename):
    
    json = _load_json(filename)
    
    data = []
    k = "duree_de_la_mission"
    data.append(f"{k}: {json[k]}")

    k = "nombre_de_profils"
    data.append(f"{k}: {json[k]}")

    k = "types_de_profils"
    data.append(f"{k}: {json[k]}")

    k = "taches_et_responsabilites"
    tmp = json[k]
    data.append(f"{k}")
    for v in tmp:
        for k1, v1 in v.items():
            data.append(f"\t{k1} : {v1}")

    k = "competences_du_profils"
    tmp = json[k]
    data.append(k)
    for v in tmp:
        for k1, v1 in v.items():
            data.append(f"\t{k1} : {v1}")
            
    return data
```

```python
cdc_data = cdc_read(cdc_json_name)

for s in cdc_data:
    print(s)
```

# CV

## Expériences

```python
cv_exp_json_string = _load_json(cv_exp_json_name)

print(json.dumps(cv_exp_json_string, indent=4))
```

```python
def cv_exp_read(filename):
    data = _load_json(filename)

    xp = data["experiences_entreprise"]

    data = []
    for x in xp:
        for k, v in x.items():
            if k == "nom_entreprise":
                data.append(", ".join([str(x) for x in v]))
            else:
                data.append(f"\t{k} : {v}")

    return data
```

```python
cv_exp_data = cv_exp_read(cv_exp_json_name)

for s in cv_exp_data:
    print(s)
```

## Taches

```python
cv_taches_json_string = _load_json(cv_taches_json_name)

print(json.dumps(cv_taches_json_string, indent=4))
```

```python
def cv_tasks_read(filename):
    json = _load_json(filename)

    xp = json["experiences_entreprise"]

    data = []
    for x in xp:
        for k, v in x.items():
            if k == "nom_entreprise":
                data.append(str(v))
            else:
                data.append(f"\t{k} : {v}")

    return data
```

```python
cv_taches_data = cv_tasks_read(cv_taches_json_name)

for s in cv_taches_data:
    print(s)
```

# Conversion json en excel

```python
def explode_rows(df, row_name, attribute_name=None):
    rows = df[df['index'] == row_name]
    exploded = rows.explode('data')
    
    if attribute_name:
        exploded['data'] = exploded['data'].apply(lambda x: x[attribute_name])
    
    dg = df.drop(df[df['index'] == row_name].index)
    dg = pd.concat([dg, exploded], ignore_index=True)
    
    return dg
```

```python
rec_df = pd.json_normalize(rec_json_string).T.reset_index()
rec_df.columns = ['index', 'data']

rec_df = explode_rows(rec_df, row_name="types_de_profils", attribute_name="type")
rec_df = explode_rows(rec_df, row_name="types_de_profils_variante", attribute_name="type")
rec_df = explode_rows(rec_df, row_name="offre_technique", attribute_name="element")
rec_df = explode_rows(rec_df, row_name="sous_criteres_techniques_et_ponderation")
rec_df['categorie'] = 'REC'
rec_df = rec_df[['categorie', 'index', 'data']]

rec_df
```

## Cahier des charges

```python
cdc_df = pd.json_normalize(cdc_json_string).T.reset_index()
cdc_df.columns = ['index', 'data']

cdc_df = explode_rows(cdc_df, row_name="types_de_profils", attribute_name="type")
cdc_df = explode_rows(cdc_df, row_name="taches_et_responsabilites", attribute_name="tache_et_responsabilite")
cdc_df = explode_rows(cdc_df, row_name="competences_du_profils", attribute_name="competence")

cdc_df['categorie'] = 'CDC'
cdc_df = cdc_df[['categorie', 'index', 'data']]

cdc_df
```

## CV

### Experiences

```python
cv_exp_df = pd.DataFrame(columns=['categorie', 'index', 'data'])
for xp in cv_exp_json_string["experiences_entreprise"]:
    tmp_df = pd.json_normalize(xp).T.reset_index()
    tmp_df.columns = ['index', 'data']

    categorie_name = tmp_df[tmp_df['index'] == "nom_entreprise"]["data"][0][0]
    tmp_df['categorie'] = categorie_name
    
    cv_exp_df = pd.concat([cv_exp_df, tmp_df], ignore_index=True)
    
cv_exp_df
```

## Tâches et responsabilités

```python
cv_taches_df = pd.DataFrame(columns=['categorie', 'index', 'data'])
for xp in cv_taches_json_string["experiences_entreprise"]:
    tmp_df = pd.json_normalize(xp).T.reset_index()
    tmp_df.columns = ['index', 'data']
    
    tmp_dg = tmp_df[tmp_df['index'] == 'role'].explode(column='data')
    tmp_df = tmp_df.drop(tmp_df[tmp_df['index'] == 'role'].index)
    tmp_df = pd.concat([tmp_df, tmp_dg], ignore_index=True)

    categorie_name = tmp_df[tmp_df['index'] == "nom_entreprise"]["data"][0][0]
    tmp_df['categorie'] = categorie_name
    
    cv_taches_df = pd.concat([cv_taches_df, tmp_df], ignore_index=True)
    
cv_taches_df
```

## Diplômes 

```python
cv_diplomes_df = pd.json_normalize(cv_diplomes_json_string).T.reset_index()
cv_diplomes_df.columns = ['index', 'data']

cv_diplomes_df = cv_diplomes_df.explode(column='data')
cv_diplomes_df['categorie'] = "Divers (diplomes, ...)"
cv_diplomes_df = cv_diplomes_df[['categorie', 'index', 'data']]

cv_diplomes_df
```

## Merge

```python
# AO SNCF
ao_df = pd.concat([rec_df, cdc_df], ignore_index=True)

## Missions
cv_df = pd.DataFrame(columns=['categorie', 'index', 'data'])

entreprises = cv_exp_df['categorie'].unique()

for e in entreprises:
    df_1 = cv_exp_df[cv_exp_df['categorie'] == e]
    
    df_2 = cv_taches_df[cv_taches_df['categorie'] == e]
    df_2 = df_2.drop(df_2[df_2['index'] == 'nom_entreprise'].index)
    df_2 = df_2.drop(df_2[df_2['index'] == 'nom_poste'].index)
    
    df_1 = pd.concat([df_1, df_2], ignore_index=True)
    
    cv_df = pd.concat([cv_df, df_1], ignore_index=True)
    
cv_df = pd.concat([cv_df, cv_diplomes_df], ignore_index=True)
```

```python
# Save to excel
writer = pd.ExcelWriter(xlsx_outname, engine='xlsxwriter')

workbook = writer.book
formats = workbook.add_format({'text_wrap': True})

ao_df.to_excel(writer, sheet_name='ao_sncf', index=False)
for column in ao_df:
    col_idx = ao_df.columns.get_loc(column)
    if column == 'data':
        writer.sheets['ao_sncf'].set_column(col_idx, col_idx, 60, formats)
    else:
        writer.sheets['ao_sncf'].set_column(col_idx, col_idx, 30, formats)

cv_df.to_excel(writer, sheet_name='cv_1', index=False)
for column in cv_df:
    col_idx = cv_df.columns.get_loc(column)
    if column == 'data':
        writer.sheets['cv_1'].set_column(col_idx, col_idx, 60, formats)
    else:
        writer.sheets['cv_1'].set_column(col_idx, col_idx, 30, formats)

writer.close()
```
